#### Weather Feature Correlation

```python
weather_df = df[['Temperature(F)', 'Wind_Chill(F)', 'Humidity(%)', 'Pressure(in)',
       'Visibility(mi)', 'Wind_Speed(mph)',
       'Precipitation(in)']]
corr = weather_df.corr()
sns.heatmap(corr, 
        xticklabels=corr.columns,
        yticklabels=corr.columns)
```

#### Missing Values

```python
msno.bar(df) 
```

```python
total = df.isnull().sum().sort_values(ascending=False)
percent = (df.isnull().sum()/df.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data.head(20)
```

```python
msno.heatmap(df) 
```

Heatmap shows the correlation of missingness between every 2 columns.
A value near -1 means if one variable appears then the other variable is very likely to be missing.
A value near 0 means there is no dependence between the occurrence of missing values of two variables.
A value near 1 means if one variable appears then the other variable is very likely to be present.

#### Correlation Among Weather Features

```python
sns.set()
cols = ['Temperature(F)', 'Humidity(%)', 'Pressure(in)', 'Visibility(mi)', 'Wind_Speed(mph)',
       'Precipitation(in)', 'Severity']
sns.pairplot(df[cols], size = 2.5)
plt.show();
```

#### Distribution of Weather Features

```python
cols = ['Precipitation(in)', 'Wind_Speed(mph)', 'Visibility(mi)', 'Pressure(in)', 'Temperature(F)']
for col in cols:
    sns.boxplot(x="variable", y="value", data=pd.melt(df[[col]]))
    plt.show()
```

#### Understanding Distribution of Categorical features

```python
bool_feat = df.select_dtypes(include='bool').columns
cat_feat = df.select_dtypes(include='object').columns
for col in cat_feat:
    print (col)
    print (df[col].value_counts(), '\n')
```
